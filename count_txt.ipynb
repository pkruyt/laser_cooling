```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from the text file, ignoring non-numeric lines
data = []
with open('SPS_pulsed.txt', 'r') as file:
    for line in file:
        # Ignore lines starting with '#' (comments) and empty lines
        if not line.startswith('#') and line.strip():
            try:
                value = float(line.strip())
                data.append(value)
            except ValueError:
                # Ignore lines that cannot be converted to float
                pass

# Plot the histogram
plt.hist(data, bins=10, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Rabi frequency')
plt.show()
```
