```python
import pickle
import numpy as np
import matplotlib.pyplot  as plt
from matplotlib.widgets import Slider
import sys

PATH='laser_cooling/cache/good_results/'

sys.path.append(PATH)


# x=np.load('cache/x.npy')
# px=np.load('cache/px.npy')
# max_x_list=np.load('cache/max_x_list.npy')
# min_x_list=np.load('cache/min_x_list.npy')
# lower_bound_list=np.load('cache/lower_bound_list.npy')

x=np.load('cooling_moving_laser_non_linear(100x)/x.npy')
px=np.load('cooling_moving_laser_non_linear(100x)/px.npy')
max_x_list=np.load('cooling_moving_laser_non_linear(100x)/max_x_list.npy')
min_x_list=np.load('good_results/cooling_moving_laser_non_linear(100x)/min_x_list.npy')
lower_bound_list=np.load('cooling_moving_laser_non_linear(100x)/lower_bound_list.npy')



num_turns = len(x)

a1=-0.0010
a2=-0.0008

a1=0.0014
a2=0.0016




```

```python


sample=500

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots()
line = ax.scatter(x[0:sample], px[0:sample],color='orange',label='initial')

fig.suptitle('Energy reduction with dispersion')
ax.set_xlabel('x(m)')
ax.set_ylabel('px')
ax.axvline(x=a1,color='red',label='cooling zone')
ax.axvline(x=a2,color='red')
ax.legend()
    
fig.subplots_adjust(left=0.25, bottom=0.25)

axfreq = fig.add_axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(
    ax=axfreq,
    label='Turns',
    valmin=0,
    valmax=num_turns-sample,
    valinit=1,
)

# The function to be called anytime a slider's value changes
# def update(val):
#     value=freq_slider.val
    
#     line.set_xdata(x[value:500])
#     line.set_ydata(px[value.val:500])
#     fig.canvas.draw_idle()


def update(val):
    value=freq_slider.val
    ax.clear()
    
    a1=lower_bound_list[int(value)]
    a2=max_x_list[int(value)]
    
    ax.axvline(x=a1,color='red',label='cooling zone')
    ax.axvline(x=a2,color='red')
    ax.scatter(x[0:sample], px[0:sample],color='orange',label='initial')
    ax.scatter(x[int(value):int(value)+sample],px[int(value):int(value)+sample]
               ,label='turn evolution')
    ax.legend()


    #plt.draw()

    
freq_slider.on_changed(update)
```
