```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the one-dimensional diffusion equation
def diffusion_equation(t, u, D):
    du_dx2 = np.gradient(np.gradient(u))
    return D * du_dx2

# Define parameters
D = 1.0  # Diffusion coefficient
L = 10.0  # Length of the spatial domain
T = 100.0  # Total simulation time
Nx = 100  # Number of spatial points
x = np.linspace(0, L, Nx)
u0 = np.exp(-(x - L / 2)**2 / (2 * 1.0**2)) / np.sqrt(2 * np.pi * 1.0**2)  # Initial condition

# Solve the diffusion equation using solve_ivp
sol = solve_ivp(diffusion_equation, (0, T), u0, args=(D,), t_eval=np.linspace(0, T, Nx))

# Plot the results
plt.imshow(sol.y, extent=[0, T, 0, L], aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(label='Quantity')
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('One-Dimensional Diffusion Equation Numerical Solution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


# Plot the results
plt.figure(figsize=(12, 5))
plt.plot(x, u0, label='Initial Condition')
plt.plot(x, sol.y[:, -1], label='Final Condition', color='orange')
plt.title('Initial Condition')
plt.xlabel('Position')
plt.ylabel('Quantity')
plt.legend()



```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.stats import norm

# Define the one-dimensional diffusion equation with a Gaussian pulse
def diffusion_equation(t, u, D, mu, sigma):
    du_dx2 = np.gradient(np.gradient(u))
    
    # Add a Gaussian pulse centered at 5
    gaussian_pulse = norm.pdf(x, loc=mu, scale=sigma)
    
    return D * du_dx2 + gaussian_pulse

# Define parameters
D = 0.1  # Diffusion coefficient (small value)
mu = 5.0  # Center of the Gaussian pulse
sigma = 0.01  # Standard deviation of the Gaussian pulse
L = 10.0  # Length of the spatial domain
T = 100.0  # Total simulation time
Nx = 1000  # Number of spatial points
x = np.linspace(0, L, Nx)
u0 = np.exp(-(x - L / 2)**2 / (2 * 1.0**2)) / np.sqrt(2 * np.pi * 1.0**2)  # Initial condition

# Solve the diffusion equation using solve_ivp
sol = solve_ivp(diffusion_equation, (0, T), u0, args=(D, mu, sigma), t_eval=np.linspace(0, T, Nx), method='RK45')

# Plot the results
plt.imshow(sol.y, extent=[0, T, 0, L], aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(label='Quantity')
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('One-Dimensional Diffusion Equation with Gaussian Pulse')
plt.show()
```
